```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

from sacrebleu.metrics import BLEU
from rouge import Rouge
```

```python
FILES = [
    "output_llama-7b_q4_0.csv",
    "output_llama-7b_q8_0.csv",
    "output_llama-13b_q4_0.csv",
    "output_vicuna-7b-v1.3_q4_0.csv",
    "output_vicuna-7b-v1.3_q8_0.csv",
    "output_vicuna-13b-v1.3_q4_0.csv"
]
```

```python
for file in FILES:
    df = pd.read_csv(file)
    display(df)
    
    bleu_scorer = BLEU(effective_order=True)

    print(file)
    print("BLEU score:", df.apply(lambda x: bleu_scorer.sentence_score(hypothesis=x[2].lower(),references=[x[1]]).score/100, axis=1).mean())
    print("ROUGE-L score:", df.apply(lambda x: Rouge().get_scores(hyps=x[2].lower(),refs=x[1])[0]["rouge-l"]["f"], axis=1).mean())
```
